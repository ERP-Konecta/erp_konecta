```python
standard_scaler = StandardScaler()
X_train_scaled = standard_scaler.fit_transform(X_train)
X_test_scaled = standard_scaler.transform(X_test)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_scaled_df = pd.DataFrame(X_train_scaled, columns=X.columns)

vif_data = pd.DataFrame()
vif_data["feature"] = X_scaled_df.columns
vif_data["VIF"] = [variance_inflation_factor(X_scaled_df.values, i)
                   for i in range(len(X_scaled_df.columns))]

print(vif_data.sort_values(by='VIF', ascending=False))
```

* VIF works by calculating a "redundancy score" for every single feature. It checks each feature (like Age) and sees how well it could be predicted by all the other features in the model.
* As the value increase the correlation between that attribute and other attributes increase as well which refelcts it may be redundant and can be removed.
* The INF values here mean that these attributes are redundant and can be compeletly known from other attributes and they introduce multicolinearity to our model.
* We can safely remove these attributes from our features set input to our models.
* This cell was run multiple times, attributes that had a score >10 were dropped from the features set and shown here are the stable features set that are all independant of each other avoiding multicolinearity.

```python
from sklearn.utils import resample
df_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
df_train_scaled['Attrition'] = y_train.values

df_majority = df_train_scaled[df_train_scaled.Attrition == 0]
df_minority = df_train_scaled[df_train_scaled.Attrition == 1]

df_majority_undersampled = resample(df_majority,
                                 replace=False,
                                 n_samples=500,
                                 random_state=42)

df_minority_oversampled = resample(df_minority,
                                 replace=True,
                                 n_samples=250,
                                 random_state=42)

df_train_resampled = pd.concat([df_majority_undersampled, df_minority_oversampled])

df_train_resampled = df_train_resampled.sample(frac=1, random_state=42)

X_train_resampled = df_train_resampled.drop('Attrition', axis=1)
y_train_resampled = df_train_resampled['Attrition']
```

Due to class imbalance, several attempts were made to reach the best compromise to solve that issue as follows.


1.  Tried to run the model with the imbalance.
2.  Tried to oversample the minority class to get both classes to have 986 values.
3.  Tried to take 500 samples from the majority class and oversample the minority class to 500 samples as well.
4.  Tried to take 500 samples from the majority class and 250 samples from the minority class.

Decided to go with the last option since it had the best performance and blanace(not to have bias nor overfitting).



```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_resampled, y_train_resampled)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(classification_report(y_test, y_pred))
print(cm)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
import tensorflow as tf # Assuming you've imported this for the NN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt



model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape=(X_train_resampled.shape[1],)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

early_stopper = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history = model.fit(X_train_resampled, y_train_resampled,
                    epochs=100,
                    batch_size=32,
                    validation_data=(X_test_scaled, y_test),
                    callbacks=[early_stopper])

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Checking for Overfitting')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

y_pred_nn = model.predict(X_test_scaled)
y_pred_nn = (y_pred_nn > 0.5).astype("int32")
cm_nn = confusion_matrix(y_test, y_pred_nn)

print(classification_report(y_test, y_pred_nn))
print(cm_nn)
```

```python
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import pandas as pd



dtrain = xgb.DMatrix(X_train_resampled, label=y_train_resampled)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_train_resampled.columns)


dval = xgb.DMatrix(X_test_scaled_df, label=y_test)
dtest = xgb.DMatrix(X_test_scaled_df)

params = {
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'random_state': 42
}

evals_result = {}

bst = xgb.train(
    params,
    dtrain,
    num_boost_round=200,
    evals=[(dtrain, 'train'), (dval, 'validation')],
    early_stopping_rounds=10,
    evals_result=evals_result,
    verbose_eval=False
)

train_loss = evals_result['train']['logloss']
val_loss = evals_result['validation']['logloss']

plt.figure(figsize=(10, 6))
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('XGBoost with Early Stopping (Core API)')
plt.xlabel('Number of Trees')
plt.ylabel('LogLoss')
plt.legend()
plt.show()

y_pred_prob = bst.predict(dtest, iteration_range=(0, bst.best_iteration))
y_pred_xgb = (y_pred_prob > 0.5).astype(int)
cm_xgb = confusion_matrix(y_test, y_pred_xgb)

print(classification_report(y_test, y_pred_xgb))

print(cm_xgb)
```

```python
# This is the master prompt. You can tune this to change the output.
PROMPT_TEMPLATE = """
You are an expert HR Analyst. Your task is to analyze employee data to identify training gaps and recommend actions.

Based on the employee data provided below, perform the following 4-step reasoning:

**Employee Data:**
{employee_data_string}

---
**Reasoning and Analysis:**

**1. Identify the Gap:**
* First, state the primary performance concern. Is there a gap based on `ManagerRating <= 3` or `SelfRating <= 3`?
* Identify if this is a high attrition risk based on the `Attrition = 1` flag.

**2. Synthesize Context (The "Why"):**
* Analyze the relationship between performance (`ManagerRating`, `SelfRating`) and satisfaction (`JobSatisfaction`, `EnvironmentSatisfaction`). Are they aligned or conflicting?
* Examine the training history: `TrainingOpportunitiesWithinYear` vs. `TrainingOpportunitiesTaken`. Is there a missed opportunity?
* Look at workload and tenure: How might `OverTime`, `YearsAtCompany`, and `YearsSinceLastPromotion` be contributing to the performance gap?
* Consider the `JobRole`: What specific skills might be lacking for this role given the data?

**3. Identified Gaps (Summary List):**
* Based on your synthesis in Step 2, provide a clear, bulleted list of the specific skill or knowledge gaps this employee appears to have. (e.g., "Technical Product Knowledge," "Time Management," "Client Negotiation").

**4. Actionable Recommendations:**
* Based on the gaps identified in Step 3, recommend 2-3 specific, actionable steps for the employee's manager.
* Suggest specific training topics or interventions for each identified gap.

Format your output clearly with these four bolded headings.
"""
```
